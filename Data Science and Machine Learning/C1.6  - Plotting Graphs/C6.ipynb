4. Plot a stacked bar graph depicting bronze, silvers and gold medals by China, India, Japan and Pakistan in last Asian Games.

```python
import matplotlib.pyplot as plt

Medal_Type = ['Bronze', 'Silver', 'Gold']
countries = ['China', 'India', 'Japan', 'Pakistan']

# Placeholder data for medal counts by country and type
medal_counts = {
    'China': [30, 40, 50],
    'India': [15, 20, 18],
    'Japan': [25, 30, 20],
    'Pakistan': [5, 8, 3]
}

fig, ax = plt.subplots()

bottom = [0, 0, 0]  # Bottom position for stacking

for country in countries:
    medal_count = medal_counts[country]
    ax.bar(Medal_Type, medal_count, label=country, bottom=bottom)
    bottom = [b + m for b, m in zip(bottom, medal_count)]

ax.set_xlabel('Medal Type')
ax.set_ylabel('Medal Count')
ax.set_title('Medals Won by Different Countries in the Last Asian Games')
ax.legend()
plt.tight_layout()
plt.show()
```

5. Draw a scatter plot depicting year of launch and speed of the car.

```python
import matplotlib.pyplot as plt

# Sample data: Years of launch and corresponding speeds
years = [2000, 2005, 2010, 2015, 2020]
speeds = [180, 200, 220, 240, 250]

plt.figure(figsize=(8, 6))  # Set the figure size

plt.scatter(years, speeds, color='blue', marker='o', edgecolors='black')
plt.title('Car Speed vs Year of Launch')
plt.xlabel('Year of Launch')
plt.ylabel('Speed (km/h)')
plt.grid(True)

for i, (year, speed) in enumerate(zip(years, speeds)):
    plt.annotate(f'({year}, {speed})', (year, speed), textcoords="offset points", xytext=(0,10), ha='center')

plt.tight_layout()
plt.show()
```

7. Draw a scatter plot representing Unemployment Rate vs Stock Index Price. These two data sets should specified using pandas data frame.

8. Write a Python program to draw a scatter plot for three different groups comparing weights and heights.

```python
import matplotlib.pyplot as plt
import random

# Sample data for three groups: Group A, Group B, and Group C
group_names = ['Group A', 'Group B', 'Group C']

group_data = {
    'Group A': {'weights': [random.uniform(50, 80) for _ in range(20)],
                'heights': [random.uniform(150, 190) for _ in range(20)]},
    
    'Group B': {'weights': [random.uniform(55, 85) for _ in range(20)],
                'heights': [random.uniform(155, 195) for _ in range(20)]},
    
    'Group C': {'weights': [random.uniform(60, 90) for _ in range(20)],
                'heights': [random.uniform(160, 200) for _ in range(20)]}
}

plt.figure(figsize=(10, 8))  # Set the figure size

for group_name, data in group_data.items():
    plt.scatter(data['weights'], data['heights'], label=group_name)

plt.title('Comparison of Weights and Heights for Different Groups')
plt.xlabel('Weight')
plt.ylabel('Height')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```

19. Compare the regression plots obtained by posing a query for dataset 1 and 2 in the anscombe data set while plotting for x abd y values.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Anscombe dataset
anscombe = sns.load_dataset('anscombe')

# Subset Dataset 1 and Dataset 2
dataset_1 = anscombe[anscombe['dataset'] == 'I']
dataset_2 = anscombe[anscombe['dataset'] == 'II']

# Create subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))

# Regression plot for Dataset 1
sns.regplot(data=dataset_1, x='x', y='y', ax=axes[0], scatter_kws={'alpha': 0.7})
axes[0].set_title('Dataset 1')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].grid(True)

# Regression plot for Dataset 2
sns.regplot(data=dataset_2, x='x', y='y', ax=axes[1], scatter_kws={'alpha': 0.7})
axes[1].set_title('Dataset 2')
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')
axes[1].grid(True)

plt.tight_layout()
plt.show()
```

21. Draw a PairGrid for the iris data set. Customize color of the plot using a categorical variable (like species). Also use a different function on the diagonal to show the univariate distribution of the variable in each column.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = sns.load_dataset('iris')

# Create a PairGrid with scatter plots
g = sns.PairGrid(iris, hue='species', diag_sharey=False, height=2.5)

# Scatter plots on the off-diagonal
g.map_upper(sns.scatterplot, alpha=0.8)

# KDE plots on the diagonal
g.map_diag(sns.kdeplot, lw=2, legend=False)

# Customize legend
g.add_legend()

plt.suptitle('PairGrid: Pair-wise Scatter Plot with Diagonal KDEs for the Iris Dataset', y=1.02)
plt.tight_layout()
plt.show()
```
