Write a Python code to detect outliers using Z-score technique.

```python
import pandas as pd
from scipy.stats import zscore
```

```python
# Load the CSV file
data = pd.read_csv('car-mpg.csv')
```

```python
# Calculate Z-scores for each column
z_scores = zscore(data)
z_scores
```

```python
# Create a DataFrame to hold Z-scores
z_scores_df = pd.DataFrame(z_scores, columns=data.columns)
z_scores_df
```

```python
# Set the Z-score threshold for outlier detection
z_score_threshold = 3  # You can adjust this threshold as needed
```

```python
# Identify outliers based on Z-scores
outliers = (z_scores_df > z_score_threshold) | (z_scores_df < -z_score_threshold)
```

```python
# Print the indices of outlier rows for each column
for column in outliers.columns:
    column_outliers = outliers.index[outliers[column]]
    print(f"Outliers in column '{column}': {column_outliers}")
```
