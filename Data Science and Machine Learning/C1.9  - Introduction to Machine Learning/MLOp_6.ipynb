Use the diabetes.csv file to rescale the data values using MinMaxScaler in the range of 0 to 1. Print the first 10 rows in the output.

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

# Load the CSV file
file_path = 'Diabetes.csv'
data = pd.read_csv(file_path)

# Select the columns you want to rescale
columns_to_rescale = data.columns

# Initialize the MinMaxScaler
scaler = MinMaxScaler()

# Fit and transform the selected columns
data[columns_to_rescale] = scaler.fit_transform(data[columns_to_rescale])

# Print the first 10 rows of the rescaled data
print(data.head(10))
```

Read the diabetes.csv and then with the help of StandardScaler class, convert the data into Gaussian Distribution with mean = 0 and SD = 1.

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler

# Load the CSV file
file_path = 'Diabetes.csv'
data = pd.read_csv(file_path)

# Select the columns you want to transform
columns_to_transform = data.columns

# Initialize the StandardScaler
scaler = StandardScaler()

# Fit and transform the selected columns
data[columns_to_transform] = scaler.fit_transform(data[columns_to_transform])

# Print the first 10 rows of the transformed data
print(data.head(10))
```
